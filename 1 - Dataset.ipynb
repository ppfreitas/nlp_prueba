```python
import pandas as pd

from utils.preproc import *
```

## Import dataset

```python
# import dataset
df = pd.read_csv('set_de_datos_con_perspectiva_de_genero - set_de_datos_unificado.csv')
df = df[0:3411] #only 3411 rows with data
df
```

```python
# fill missing values for N with previous value
df['N'].fillna(method='ffill', inplace=True)

# remove rows with missing links as the docs are the feature of our model
df = df[~df.LINK.isnull()]
df.shape
```

## Download docs, extract texts, and save to csv

```python
# extract google drive ids from links
ids = [get_id(url) for url in df['LINK']]
```

```python
# extract text given id
texts = []
i = 0
for doc_id in ids:
    print(i)
    try:
        text = get_text2(doc_id) #download doc file and extract its text
        text = text.decode('utf-8')
    except:
        text = None
        print('Download failed')
        
    texts.append(text)
    i = i+1
```

```python
# add ids and texts to df
df.loc[:,'DOCS_ID'] = ids
df.loc[:,'TEXTOS'] = texts
```

```python
df.shape
```

```python
# missing texts. 
df.TEXTOS.isnull().sum()
```

```python
# save df generated to csv to be used in the steps
df.to_csv('df_raw_text2.csv')
```
